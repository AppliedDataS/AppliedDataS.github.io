```python
def generateNNv(layers, edge_label_on=True,
                color=['lightblue', 'lightpink', 'lightgreen'],
                node_size=1100,
                label='Neural Network',
                edgesep='.002',
                fillcolor='none',
                fontcolor='Black',
                fontname='Arial',
                fontsize='32',
                labelloc="t",
                line='none',
                nodesep='.21',
                pencolor='none',
                penwidth='0',
                rankdir='LR',
                ranksep="2.0",
                splines="line",
                style='filled',
                node_color='b',
                node_edgesep='.002',
                node_fillcolor='none',
                node_fontcolor='blue',
                node_fontname='Arial',
                node_fontsize='32',
                node_labelloc="t",
                node_line='none',
                node_nodesep='.21',
                node_pencolor='none',
                node_penwidth='0',
                node_rankdir='LR',
                node_ranksep="2.0",
                node_shape='circle',
                node_splines="line",
                node_style='filled',
                edge_arrowsize='.5',
                edge_color='darkviolet',
                edge_edgesep='.002',
                edge_fillcolor='none',
                edge_fontcolor='Black',
                edge_fontname='Arial',
                edge_fontsize='12',
                edge_labelloc="t",
                edge_line='none',
                edge_nodesep='.21',
                edge_pencolor='darkgreen',
                edge_penwidth='2',
                edge_rankdir='LR',
                edge_ranksep="2.0",
                edge_splines="line",
                edge_style='filled',
                subgraph_label='',
                subgraph_edgesep='.002',
                subgraph_fillcolor='none',
                subgraph_fontcolor='Black',
                subgraph_fontname='Arial',
                subgraph_fontsize='16',
                subgraph_labelloc="b",
                subgraph_line='none',
                subgraph_nodesep='.21',
                subgraph_pencolor='none',
                subgraph_penwidth='0',
                subgraph_rankdir='LR',
                subgraph_ranksep="2.0",
                subgraph_splines="line",
                subgraph_style='filled',
                subgraph_node_color='b', edge_label='w'):

    layers_str = ["Input"] + ["Hidden"] * (len(layers) - 2) + ["Output"]
    layers_col = ["none"] + ["none"] * (len(layers) - 2) + ["none"]
    layers_fill = [color[0]] + [color[1]] * (len(layers) - 2) + [color[-1]]
    g = Digraph(label)
    g.attr(label=label)
    g.attr(fontsize=fontsize)
    g.attr(fontname=fontname)
    g.attr(fontcolor=fontcolor)
    g.attr(style=style)
    g.attr(fillcolor=fillcolor)
    g.attr(penwidth=penwidth)
    g.attr(pencolor=pencolor)
    g.attr(rankdir=rankdir)
    g.attr(line=line)
    g.attr(splines=splines)
    g.attr(nodesep=nodesep)
    g.attr(edgesep=nodesep)
    g.attr(ranksep=ranksep)
    g.attr(labelloc=labelloc)
    g.attr('node', fontname=node_fontname)
    g.attr('edge', color=edge_color, arrowsize=edge_arrowsize)
    n = len(layers)

    for i in range(1, n+1):
        for j in range(layers[i-1]):
            g.attr('node',
                   shape=node_shape,
                   style=node_style,
                   color=layers_fill[i-1],
                   fillcolor=layers_fill[i-1],
                   fontcolor=node_fontcolor,
                   rankdir=node_rankdir)
            g.node(f'a{i}{j}', label=f'a{j}({i-1})')

    for i in range(1, n):
        h = layers[i - 1]
        for a in range(h):
            for b in range(layers[i]):
                label = f'{edge_label}{i-1}{i}'
                g.attr('edge', pencolor=edge_pencolor, penwidth=edge_penwidth)
                if edge_label_on:
                    g.attr('edge', label=label)
                g.edge(f'a{i}{a}', f'a{i+1}{b}')

    for i in range(1, n+1):
        # nodes
        names = []
        for j in range(layers[i-1]):
            g.attr('node',
                   color=layers_fill[i-1],
                   fillcolor=layers_fill[i-1],
                   )
            g.node(f'a{i}{j}', label=f'a{j}({i-1})')

    for i in range(1, n+1):
        with g.subgraph(name=f'cluster_{i}') as c:
            c.attr(fontsize=subgraph_fontsize)
            c.attr(labelloc=subgraph_labelloc)
            for j in range(layers[i-1]):
                c.attr(label=layers_str[i-1], fontcolor=subgraph_fontcolor,
                       fillcolor=subgraph_fillcolor, penwidth=subgraph_penwidth)
                c.node(f'a{i}{j}')

    return g
```

```python
fig, ax = plt.subplots(figsize=(1, 1))
ax.set_axis_off()
fig.tight_layout()
tex = "$a^{(1)} = \sigma(\omega_{10}{a^{(0)}}+b_0)$\n"
tex += "where:\n"
tex += "$a$: activity\n"
tex += "$\sigma$: activation function\n"
tex += "$\omega$: weight\n"
tex += "$b$: bias.\n"
text = ax.text(0, 0, tex, c='darkblue',
                fontfamily='arial', fontsize='16')
generateNNv([1,1])
```
